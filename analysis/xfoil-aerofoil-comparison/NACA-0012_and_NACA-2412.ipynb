```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both CSVs - comma separated, skip 9 header lines
df0012 = pd.read_csv('T1_Re1.000_M0.00_N9.0_NACA-0012.csv', skiprows=9,
                     names=['alpha','Cl','Cd','CDp','Cm','Top_Xtr','Bot_Xtr',
                            'Cpmin','Chinge','XCp','extra1','extra2'])

df2412 = pd.read_csv('T1_Re1.000_M0.00_N9.0_NACA-2412.csv', skiprows=9,
                     names=['alpha','Cl','Cd','CDp','Cm','Top_Xtr','Bot_Xtr',
                            'Cpmin','Chinge','XCp','extra1','extra2'])

# Convert to numeric just in case
df0012 = df0012.apply(pd.to_numeric, errors='coerce')
df2412 = df2412.apply(pd.to_numeric, errors='coerce')

# Plot 1 — Cl vs Alpha
plt.figure(figsize=(8,5))
plt.plot(df0012['alpha'], df0012['Cl'], 'b-o', markersize=3, label='NACA 0012 (root)')
plt.plot(df2412['alpha'], df2412['Cl'], 'g-o', markersize=3, label='NACA 2412 (tip)')
plt.axhline(0, color='gray', linewidth=0.5, linestyle='--')
plt.axvline(0, color='gray', linewidth=0.5, linestyle='--')
plt.xlabel('Angle of Attack α (degrees)')
plt.ylabel('Lift Coefficient Cl')
plt.title('Cl vs α — NACA 0012 vs NACA 2412 at Re = 1,000,000')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('cl_vs_alpha_comparison.png', dpi=150)
plt.show()

# Plot 2 — Cd vs Alpha
plt.figure(figsize=(8,5))
plt.plot(df0012['alpha'], df0012['Cd'], 'b-o', markersize=3, label='NACA 0012 (root)')
plt.plot(df2412['alpha'], df2412['Cd'], 'g-o', markersize=3, label='NACA 2412 (tip)')
plt.xlabel('Angle of Attack α (degrees)')
plt.ylabel('Drag Coefficient Cd')
plt.title('Cd vs α — NACA 0012 vs NACA 2412 at Re = 1,000,000')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('cd_vs_alpha_comparison.png', dpi=150)
plt.show()

# Print key results
print("=== NACA 0012 Key Results ===")
print(f"Max Cl: {df0012['Cl'].max():.4f} at α = {df0012.loc[df0012['Cl'].idxmax(), 'alpha']:.1f}°")
print(f"Min Cd: {df0012['Cd'].min():.5f}")
zero_lift_0012 = df0012.loc[(df0012['Cl'].abs()).idxmin(), 'alpha']
print(f"Zero-lift angle: ~{zero_lift_0012:.1f}°")

print("\n=== NACA 2412 Key Results ===")
print(f"Max Cl: {df2412['Cl'].max():.4f} at α = {df2412.loc[df2412['Cl'].idxmax(), 'alpha']:.1f}°")
print(f"Min Cd: {df2412['Cd'].min():.5f}")
zero_lift_2412 = df2412.loc[(df2412['Cl'].abs()).idxmin(), 'alpha']
print(f"Zero-lift angle: ~{zero_lift_2412:.1f}°")
```
